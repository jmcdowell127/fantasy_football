```python
# import scraping modules
from urllib.request import urlopen
from bs4 import BeautifulSoup

# import pandas, numpy, and matplotlib
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
url = "https://www.pro-football-reference.com/years/2021/passing.htm"

# open url and pass to beautiful soup
html = urlopen(url)
stats_page = BeautifulSoup(html)
```

```python
# collect table headers
column_headers = stats_page.findAll('tr')[0]
column_headers = [i.getText() for i in column_headers.findAll('th')]
```

```python
print(column_headers)
```

```python
# collect table rows
rows = stats_page.findAll('tr')[1:]

# get stats from each row
qb_stats = []
for i in range(len(rows)):
    qb_stats.append([col.getText() for col in rows[i].findAll('td')])
```

```python
# look at first row of qb stats
print(qb_stats[0])
```

```python
# create dataframe
data = pd.DataFrame(qb_stats, columns=column_headers[1:])
data.head(15)
```

## Clean up Data Frame

```python
# view columns
data.columns
```

```python
# rename sack yards column to Yds_Sack
new_columns = data.columns.values
new_columns[-6] = 'Yds_Sack'
data.columns = new_columns
```

```python
data.columns
```

## Create a new Data Frame for QBs

```python
# stat categories
cats = ['Yds', 'TD', 'Int', 'QBR', 'Rate','Cmp%']

# create data subset for chart and add player name and team
qb_chart = data[['Player', 'Tm'] + cats]
qb_chart.head(20)
```

```python
# check data types
qb_chart.dtypes
```

```python
# convert data to numerical values
for i in cats:
    qb_chart[i] = pd.to_numeric(data[i])
```

```python
qb_chart.dtypes
```

```python
# remove * and + for probowl and allpro
qb_chart['Player'] = qb_chart['Player'].str.replace('*', '')
qb_chart['Player'] = qb_chart['Player'].str.replace('*', '')

# filter passing yards
qb_chart_filtered = qb_chart[qb_chart['Yds'] > 1500]

# create columns with percentile rank
for i in cats:
    qb_chart_filtered[i + '_Rank'] = qb_chart_filtered[i].rank(pct=True)
    
# flip the rank for interceptions so that the QB with the most doest have the highest percentile
qb_chart_filtered['Int_Rank'] = 1 - qb_chart_filtered['Int_Rank']
```

```python
# check dataframe
qb_chart_filtered.head(10)
```

## Generate Radar Charts

```python
# set plot parameters
mpl.rcParams['font.family'] = 'DejaVu Sans'
mpl.rcParams['font.size'] = 16
mpl.rcParams['axes.linewidth'] = 0
mpl.rcParams['xtick.major.pad'] = 15

# set colors for teams using HEX codes from teamcolorcodes.com/nfl-team-color-codes
team_colors = {'ARI':'#97233f', 'ATL':'#a71930', 'BAL':'#241773', 'BUF':'#00338d', 'CAR':'#0085ca', 
               'CHI':'#0b162a', 'CIN':'#fb4f14', 'CLE':'#311d00', 'DAL':'#041e42', 'DEN':'#002244', 
               'DET':'#0076b6', 'GNB':'#203731', 'HOU':'#03202f', 'IND':'#002c5f', 'JAX':'#006778', 
               'KAN':'#e31837', 'LAC':'#002a5e', 'LAR':'#003594', 'MIA':'#008e97', 'MIN':'#4f2683', 
               'NWE':'#002244', 'NOR':'#d3bc8d', 'NYG':'#0b2265', 'NYJ':'#125740', 'OAK':'#000000', 
               'PHI':'#004c54', 'PIT':'#ffb612', 'SFO':'#aa0000', 'SEA':'#002244', 'TAM':'#d50a0a', 
               'TEN':'#0c2340', 'WAS':'#773141'}
```

```python
# set angles for radar charts using numpy.linspace(). since there are 6 categories the points will be plotted every
# 2π/6 radians, or 60 degrees. Also add offset term to adjust where labels appear
offset = np.pi/6
angles = np.linspace(0, 2*np.pi, len(cats) + 1) + offset
```

```python
# create helper function for radar charts
def create_radar_chart(ax, angles, player_data, color='blue'):
    
    # Plot data and fill with team color
    ax.plot(angles, np.append(player_data[-(len(angles)-1):], 
            player_data[-(len(angles)-1)]), color=color, linewidth=2)
    ax.fill(angles, np.append(player_data[-(len(angles)-1):], 
            player_data[-(len(angles)-1)]), color=color, alpha=0.2)
    
    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(cats)
    
    # Remove radial labels
    ax.set_yticklabels([])

    # Add player name
    ax.text(np.pi/2, 1.7, player_data[0], ha='center', va='center', 
            size=18, color=color)

    # Set axis limits
    ax.set(xlim=(0, 2*np.pi), ylim=(0, 1))

    return ax
```

```python
# create helper function to get QB data
def get_qb_data(data, team):
    return np.asarray(data[data['Tm'] == team]) [0]
```

```python
# create figure
plt.figure(figsize=(8, 8), facecolor='white')

# add subplots
ax1 = plt.subplot(221, polar=True, facecolor='#ededed')
ax2 = plt.subplot(222, polar=True, facecolor='#ededed')
ax3 = plt.subplot(223, polar=True, facecolor='#ededed')
ax4 = plt.subplot(224, polar=True, facecolor='#ededed')

# adjust space between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.5)

# get qb data
rams_data = get_qb_data(qb_chart_filtered, 'LAR')
bucs_data = get_qb_data(qb_chart_filtered, 'TAM')
chiefs_data = get_qb_data(qb_chart_filtered, 'KAN')
bills_data = get_qb_data(qb_chart_filtered, 'BUF')

# plot qb data
ax1 = create_radar_chart(ax1, angles, rams_data, team_colors['LAR'])
ax2 = create_radar_chart(ax2, angles, bucs_data, team_colors['TAM'])
ax3 = create_radar_chart(ax3, angles, chiefs_data, team_colors['KAN'])
ax4 = create_radar_chart(ax4, angles, bills_data, team_colors['BUF'])

plt.show()
```
